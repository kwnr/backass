```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data_pid = pd.read_csv("/home/armstrong/arm_ws/src/aic_core/log/values_log/log240313135751.csv")
data_sm = pd.read_csv("/home/armstrong/arm_ws/src/aic_core/log/values_log/log240313135521.csv")
```

```python
data_pid_u_start = 1586
data_sm_u_start = 2924
time_pid_u_at_start = data_pid["timestamp"][data_pid_u_start] - 1
time_pid_u_at_end = time_pid_u_at_start + 5
time_sm_u_at_start = data_sm["timestamp"][data_sm_u_start] - 1
time_sm_u_at_end = time_sm_u_at_start + 5
pid_u = data_pid[data_pid["timestamp"]>time_pid_u_at_start][data_pid["timestamp"]<time_pid_u_at_end]
sm_u = data_sm[data_sm["timestamp"]>time_sm_u_at_start][data_sm["timestamp"]<time_sm_u_at_end]

data_pid_d_start = 2495
data_sm_d_start = 4642
time_pid_d_at_start = data_pid["timestamp"][data_pid_d_start] - 1
time_pid_d_at_end = time_pid_d_at_start + 5
time_sm_d_at_start = data_sm["timestamp"][data_sm_d_start] - 1 
time_sm_d_at_end = time_sm_d_at_start + 5
pid_d = data_pid[data_pid["timestamp"]>time_pid_d_at_start][data_pid["timestamp"]<time_pid_d_at_end]
sm_d = data_sm[data_sm["timestamp"]>time_sm_d_at_start][data_sm["timestamp"]<time_sm_d_at_end]
fig, axes = plt.subplots(2, 5)
fig.set_size_inches(24,10)
fig.suptitle("Comparison between PID and +S", fontsize='xx-large')
fig.text(.1,.05,"used data: log240313135751.csv for pid, log240313135521.csv for pid+s\nstep given at 1s")
for i, ax in enumerate(axes[0]):
    if i > 1:
        i += 1
    a = f"R{i+1} pos"
    ax.plot(pid_u["timestamp"]-time_pid_u_at_start, pid_u[a])
    ax.plot(sm_u["timestamp"]-time_sm_u_at_start, sm_u[a])
    ax.set_title(a)
    ax.legend(["PID", "PID+S"])
for i, ax in enumerate(axes[1]):
    if i > 1:
        i += 1
    a = f"R{i+1} pos"
    ax.plot(pid_d["timestamp"]-time_pid_d_at_start, pid_d[a])
    ax.plot(sm_d["timestamp"]-time_sm_d_at_start, sm_d[a])
    ax.legend(["PID", "PID+S"])
```

```python
data1_start=data1["timestamp"][data1["R6 ref"] == 30].iloc[0]
data2_start=data2["timestamp"][1000:][data2["R6 ref"] == 30].iloc[0]
data3_start=data3["timestamp"][data3["R6 ref"] == 30].iloc[0]
```

```python
dt = 5
pos1 = data1["R6 pos"][data1["timestamp"]>data1_start][data1["timestamp"]<data1_start+dt]
pos2 = data2["R6 pos"][data2["timestamp"]>data2_start][data2["timestamp"]<data2_start+dt]
pos3 = data3["R6 pos"][data3["timestamp"]>data3_start][data3["timestamp"]<data3_start+dt]

time1 = data1["timestamp"][data1["timestamp"]>data1_start][data1["timestamp"]<data1_start+dt] - data1_start
time2 = data2["timestamp"][data2["timestamp"]>data2_start][data2["timestamp"]<data2_start+dt] - data2_start
time3 = data3["timestamp"][data3["timestamp"]>data3_start][data3["timestamp"]<data3_start+dt] - data3_start
```
