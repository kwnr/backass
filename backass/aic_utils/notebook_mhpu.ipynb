```python
ser = serial.Serial("/dev/ttyCUI", 2000000, timeout=.5)
```

```python
left_cui_addr = [b'\x6C', b'\x7C', b'\x8C', b'\x9C', b'\xAC', b'\xBC', b'\xEC', b'\xFC']  # L1 to 8
right_cui_addr = [b'\x0C', b'\x1C', b'\x2C', b'\x3C', b'\x4C', b'\x5C', b'\xCC', b'\xDC']  # R1 to 8
cui_addr =  left_cui_addr + right_cui_addr
```

```python
joint_name = [f"L{i}" for i in range(1,9)] + [f"R{i}" for i in range(1,9)]
for i, addr in enumerate(cui_addr):
    ser.reset_input_buffer()
    ser.write(addr)
    res = ser.read(2)
    print(joint_name[i], addr, res)
```
